# Creating Variables in R

## Variables are containers for storing data values.

## R does not have a command for declaring a variable. A variable is created the moment you first assign a value to it. To assign a value to a variable, use the <- sign. To output (or print) the variable value, just type the variable name:

```r
name <- "John"
age <- 40

name   # output "John"
age    # output 40
```

## From the example above, name and age are variables, while "John" and 40 are values.

## In other programming language, it is common to use = as an assignment operator. In R, we can use both = and <- as assignment operators.

## However, <- is preferred in most cases because the = operator can be forbidden in some context in R.

## The if Statement

An "if statement" is written with the if keyword, and it is used to specify a block of code to be executed if a condition is TRUE:

In this example we use two variables, a and b, which are used as a part of the if statement to test whether b is greater than a. As a is 33, and b is 200, we know that 200 is greater than 33, and so we print to screen that "b is greater than a".

### R uses curly brackets { } to define the scope in the code.

```r
# conditionals 
a<-as.numeric(readline(prompt="Enter the first number: "))
b<-as.numeric(readline(prompt="Enter the second number: "))
 
if (b>a){
  print(paste(b,"b is greater than a"))
} else if (a==b){
  print ("a and b are equal")
} else {
  print(paste(a,"a is greater than b"))
}
```

```r
# conditionals 
a<-as.numeric(readline(prompt="Enter the first number: "))
#b<-as.numeric(readline(prompt="Enter the second number: "))

if (a>10){
    print ('Greater than 10')
    if (a>20){
        print('Also greater than 20')
    }else{
        print("but not greater than 20")
    }
}else {
    print ("Not greater than 10")
}
```

# AND and OR

```r
# AND operator

a<-as.numeric(readline(prompt="Enter the first number: a "))
b<-as.numeric(readline(prompt="Enter the second number: b "))
c<-as.numeric(readline(prompt="Enter the third number: c "))
d<-as.numeric(readline(prompt="Enter the fourth number: d "))


if (a>b & a>c & a>d){
    print ("a is the greatest number")
}
```

```r
# OR operator

a<-as.numeric(readline(prompt="Enter the first number: a "))
b<-as.numeric(readline(prompt="Enter the second number: b "))
c<-as.numeric(readline(prompt="Enter the third number: c "))
#d<-as.numeric(readline(prompt="Enter the fourth number: d "))


if (a>b | a>c ){
    print ("a is greater than either number")
} 
```

# Loops

Loops can execute a block of code as long as a specified condition is reached.

Loops are handy because they save time, reduce errors, and they make code more readable.

# R has two loop commands:

    - while loops
    - for loops

# R While Loops

With the while loop we can execute a set of statements as long as a condition is TRUE:
Example



```r
# Print i as long as i is less than 6

n<-as.numeric(readline(prompt="Enter the ending number: "))
i<-as.numeric(readline(prompt="Enter the starting number: "))

while(i<n){
    print(i)
    i<-i+1
}
```

```r
n<-as.numeric(readline(prompt="Enter the ending number: "))
i<-as.numeric(readline(prompt="Enter the starting number: "))

while(i<n){
    print(i)
    i<-i+1
}
```

```r
n<-as.numeric(readline(prompt="Enter the ending number: "))
i<-as.numeric(readline(prompt="Enter the starting number: "))

while(i<=n){
    if (i%%2!=0){
        print(i)
    }
    i<-i+1
}
```

# R For Loop
# For Loops

### A for loop is used for iterating over a sequence:

### This is less like the for keyword in other programming languages, and works more like an iterator method as found in other object-orientated programming languages.

```r
s<-as.numeric(readline(prompt="Enter the starting number: "))
n<-as.numeric(readline(prompt="Enter the ending number: "))


for (i in s:n){
    print(i)
}
```

```r
city<-list('London','Manchester','Belfast','Nottingham')

for (c in city){
    print (c)
}
```

```r
city<-list('London','Manchester','Belfast','Nottingham')
colour<-list('Red','Blue','Cyan','Magenta')


for (c1 in city){
    for(c2 in colour){
        # concatenating the words, cities with colour
        #print(paste(c1,c2))
    }
    print(paste(c2,c1))
}
```

```r
city<-list('London','Manchester','Belfast','Nottingham','Oxford','Cambridge')
colour<-list('Red','Blue','Cyan','Magenta')

for (i in 1:length(city)){
    print(paste(city[i],colour[i]))
}
```
